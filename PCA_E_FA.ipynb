##  1.Explanação da técnica de analise


# Quem fez?
Analise de Componentes principais:  O PCA foi inventado em 1901 por Karl Pearson como uma ferramenta de Análise 
Exploratória de Dados para desenvolver modelos preditivos.

Analise fatorial: Desenvolvida em 1904 por Charles
    Spearman com o objetivo de avaliar correlações entre notas de várias 
    disciplinas cursadas por estudantes.

# Qual o propósito geral?
Analise de Componentes principais: PCA é uma técnica de 
análise multivariada que pode ser usada para analisar inter-relações
entre um grande número de variáveis e explicar essas variáveis em 
termos de suas dimensões.

Analise fatorial: Reduzir uma grande quantidade de variáveis observadas a um 
número reduzido de fatores.

# Quais sao as principais etapas?
Analise de Componentes principais:
1.   Coletar os dados de amostra;
2.   Calcular a media dos dados obtidos;
3.   Subtrair a media de todos os itens de dados;
4.   Calcular a matriz de covariância; 
5.   Calcular auto valores;
6.   Organizar a matriz de transformada;

Analise fatorial:
1.   Observar dados de entrada;
2.   Calcula as correlações entre as variáveis;
3.   Obtenção inicial dos fatores;
4.   Rotacionar a matriz;


# Algoritmos de Analise de Componentes Principais e Analise Fatorial

# Carregando Bibliotecas de desenvolvimento dos Datasets prontos e ferramentas de Plot

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn import decomposition
from sklearn.decomposition import FactorAnalysis, PCA
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_iris
```

# Carregando DataSet Iris para manipulação dos dados


```python
np.random.seed(5)

dataSetIris = load_iris()
X = dataSetIris.data
y = dataSetIris.target
```

# Separação de posições de interesse no DataSet e plot de dados por regiões de interesse

```python
figura = plt.figure(1, figsize=(4, 3))
plt.clf()

ax = figura.add_subplot(111, projection="3d", elev=48, azim=134)
ax.set_position([0, 0, 0.95, 1])
```

# Calculo das medias dos itens do DataSet e plot de grafico de cordenadas X, Y

```python
plt.cla()
pca = decomposition.PCA(n_components=3)
pca.fit(X)
X = pca.transform(X)

for name, label in [("Setosa", 0), ("Versicolour", 1), ("Virginica", 2)]:
    ax.text3D(
        X[y == label, 0].mean(),
        X[y == label, 1].mean() + 1.5,
        X[y == label, 2].mean(),
        name,
        horizontalalignment="center",
        bbox=dict(alpha=0.5, edgecolor="w", facecolor="w"),
    )
    
```

# Calculo dos auto valores e Organização da Matriz de Transformada

No algoritmo abaixo percebe-se que os itens dos ponto X possuem mais Correlações se compararmos com os pontos Y e Z. Isso porque esses dados possuem mais ligação uns com os outros.

```python
y = np.choose(y, [1, 2, 0]).astype(float)
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y, cmap=plt.cm.nipy_spectral, edgecolor="k")

ax.w_xaxis.set_ticklabels([])
ax.w_yaxis.set_ticklabels([])
ax.w_zaxis.set_ticklabels([])

plt.show()
```

```python
np.random.seed(5)

dataSetIris = load_iris()
X = dataSetIris.data
y = dataSetIris.target

figura1 = plt.figure(1, figsize=(4, 3))
plt.clf()

ax = figura1.add_subplot(111, projection="3d", elev=48, azim=134)
ax.set_position([0, 0, 0.95, 1])


plt.cla()
AnalPca = decomposition.PCA(n_components=3)
AnalPca.fit(X)
X = AnalPca.transform(X)

for name, label in [("Ponto X", 0), ("Ponto Y", 1), ("Ponto Z", 2)]:
    ax.text3D(
        X[y == label, 0].mean(),
        X[y == label, 1].mean() + 1.5,
        X[y == label, 2].mean(),
        name,
        horizontalalignment="center",
        bbox=dict(alpha=0.5, edgecolor="w", facecolor="w"),
    )
    
y = np.choose(y, [1, 2, 0]).astype(float)
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y, cmap=plt.cm.nipy_spectral, edgecolor="k")

ax.w_xaxis.set_ticklabels([])
ax.w_yaxis.set_ticklabels([])
ax.w_zaxis.set_ticklabels([])

plt.show()
```

# Analise Fatorial

Carregando DataSet para analise fatorial

```python
dataset = load_iris()
X = StandardScaler().fit_transform(dataset["data"])
feature_names = dataset["feature_names"]
```

Plot

```python
ax = plt.axes()
```

Calculo dos fatores

```python
im = ax.imshow(np.corrcoef(X.T), cmap="RdBu_r", vmin=-1, vmax=1)

ax.set_xticks([0, 1, 2, 3])
ax.set_xticklabels(list(feature_names), rotation=90)
ax.set_yticks([0, 1, 2, 3])
ax.set_yticklabels(list(feature_names))
```

Rotacionando matriz e descobrindo sua correlação

```python
plt.colorbar(im).ax.set_ylabel("$r$", rotation=0)
ax.set_title("Matriz de correlação")
plt.tight_layout()
```

Os dados analisados no DatSet em questão contém um conjunto de 150 registros com cinco atributos. Todos os calculos foram feitos com base nas 5 colunas e suas correlações.

Plot do calculo de correlação completo

```python
dataset = load_iris()
X = StandardScaler().fit_transform(dataset["data"])
feature_names = dataset["feature_names"]

ax = plt.axes()

im = ax.imshow(np.corrcoef(X.T), cmap="RdBu_r", vmin=-1, vmax=1)

ax.set_xticks([0, 1, 2, 3])
ax.set_xticklabels(list(feature_names), rotation=90)
ax.set_yticks([0, 1, 2, 3])
ax.set_yticklabels(list(feature_names))

plt.colorbar(im).ax.set_ylabel("$r$", rotation=0)
ax.set_title("Correlação da Martriz X, Y")
plt.tight_layout()
```
